```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(30) + 1
y1 = np.random.rand(30)
y2 = y1 - 5
```

---

### 折線圖

```python
plt.figure(figsize=(20, 6))
plt.title('My title', fontsize=30)
plt.plot(y1, c='red', linewidth=4, linestyle='dotted', marker='s', markersize=20, label='group1')
plt.plot(y2, c='blue', linewidth=1, linestyle='dashed', marker='*', markersize=10, label='group2')
plt.xlabel('index', fontsize=30)
plt.ylabel('value', fontsize=30)
plt.xlim(1, 30)
plt.xticks(np.arange(min(x), max(x), step=1))
plt.grid()
plt.legend(prop={'size': 20})
```
