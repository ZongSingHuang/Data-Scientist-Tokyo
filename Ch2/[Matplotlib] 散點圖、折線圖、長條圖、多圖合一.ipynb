```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(30) + 1
y1 = np.random.rand(30)
y2 = y1 - 5
```

---

### 散點圖

```python
plt.figure(figsize=(10, 3))
plt.title('My title')
plt.scatter(x, y1, c='red', marker='s', s=5, label='group1')
plt.scatter(x, y2, c='blue', marker='*', s=100, label='group2')
plt.xlabel('index')
plt.ylabel('value')
plt.xlim(1, 30)
plt.xticks(np.arange(min(x), max(x), step=2))
plt.grid()
plt.legend(prop={'size': 20})
```

---

### 折線圖

```python
plt.figure(figsize=(20, 6))
plt.title('My title', fontsize=30)
plt.plot(y1, c='red', linewidth=4, linestyle='dotted', marker='s', markersize=20, label='group1')
plt.plot(y2, c='blue', linewidth=1, linestyle='dashed', marker='*', markersize=10, label='group2')
plt.xlabel('index', fontsize=30)
plt.ylabel('value', fontsize=30)
plt.xlim(1, 30)
plt.xticks(np.arange(min(x), max(x), step=1))
plt.grid()
plt.legend(prop={'size': 20})
```

---

### 長條圖

```python
plt.figure(figsize=(20, 6))
plt.title('My title', fontsize=30)
plt.hist(y1, color='red', label='group1')
plt.hist(y2, color='blue', label='group2')
plt.xlabel('value', fontsize=30)
plt.ylabel('number of value', fontsize=30)
plt.grid()
plt.legend(prop={'size': 20})
```

---

### 多圖合一

```python
plt.figure(figsize=(20, 6))
```

```python
plt.subplot(2, 2, 1)
plt.title('My title1')
plt.scatter(x, y1, c='red', marker='s', label='group1')
plt.xlabel('index')
plt.ylabel('value')
plt.xlim(1, 30)
plt.xticks(np.arange(min(x), max(x), step=2))
plt.grid()
plt.legend()

plt.subplot(2, 2, 2)
plt.title('My title2')
plt.plot(y2, c='blue', linestyle='dashed', marker='*', label='group2')
plt.xlabel('index')
plt.ylabel('value')
plt.xlim(1, 30)
plt.xticks(np.arange(min(x), max(x), step=4))
plt.grid()
plt.legend()

plt.subplot(2, 2, 3)
plt.title('My title3')
plt.plot(y1, c='red', linestyle='dotted', marker='s', label='group3')
plt.xlabel('index')
plt.ylabel('value')
plt.xlim(1, 30)
plt.xticks(np.arange(min(x), max(x), step=6))
plt.grid()
plt.legend()

plt.subplot(2, 2, 4)
plt.title('My title4')
plt.hist(y2, color='blue', label='group4')
plt.xlabel('value')
plt.ylabel('number of value')
plt.grid()
plt.legend()


plt.tight_layout()
```
