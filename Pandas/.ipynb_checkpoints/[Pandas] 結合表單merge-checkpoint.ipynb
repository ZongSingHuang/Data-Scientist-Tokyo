```python
import numpy as np
import pandas as pd
```

```python
raw_data1 = {'ID'        :['100',     '101',   '102',   '103',      '104'],
             'City'      :['Tokyo',   'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo'],
             'Birth_year':[1990,      1989,    1992,    1997,       1982],
             'Name'      :['Hiroshi', 'Akiki', 'Yuki',  'Satoru',   'Steve']}
table1 = pd.DataFrame(raw_data1, index=['a', 'b', 'c', 'd', 'e'])
table1
```

```python
raw_data2 = {'ID'       :['100', '101', '102', '105', '107'],
             'Math'     :[50,    43,    33,    76,    98],
             'English'  :[90,    30,    20,    50,    30],
             'Sex'      :['M',   'F',   'F',   'M',   'M'],
             'Index_num':[0,     1,     2,     3,     4]}
table2 = pd.DataFrame(raw_data2)
table2
```

### 結合(merge)

```python
pd.merge(table1, table2)
```

### 內部結合(inner)

```python
pd.merge(table1, table2, on='ID')
```

### 外部結合/全結合(outer)

```python
pd.merge(table1, table2, how='outer')
```

### 左外部結合(left)

```python
pd.merge(table1, table2, how='left')
```

### 右外部結合(right)

```python
pd.merge(table1, table2, how='right')
```
