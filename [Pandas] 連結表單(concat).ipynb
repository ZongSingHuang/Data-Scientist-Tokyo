```python
import numpy as np
import pandas as pd
```

```python
raw_data1 = np.arange(16).reshape(4, 4)
table1 = pd.DataFrame(data=raw_data1)
table1
```

```python
raw_data2 = np.zeros([4, 2]) - 666
table2 = pd.DataFrame(raw_data2)
table2
```

### 堆疊

```python
pd.concat([table1, table2], axis=0).reset_index(drop=True)
```

### 並排

```python
pd.concat([table1, table2], axis=1)
```
