```python
import numpy as np
import pandas as pd
```

```python
raw_data = np.random.uniform(size=[10, 4])
table = pd.DataFrame(data=raw_data)

table
```

```python
table.iloc[1, 0] = np.nan
table.loc[5:, 3] = np.nan
table.at[2, 2] = np.nan

table
```

### nan/null 判定

```python
table.isnull()
```

```python
table[0].isnull()
```

```python
table.isnull().sum()
```

### 刪除dropna

```python
table.dropna(axis=0)
```

```python
table.dropna(axis=1)
```

```python
table[[0, 1]].dropna(axis=0)
```

### 填補fillna

補0

```python
table.fillna(0)
```

拿前一筆補

```python
table.fillna(method='ffill')
```

拿後一筆補

```python
table.fillna(method='bfill')
```

拿平均值補

```python
table.fillna(table.mean())
```
